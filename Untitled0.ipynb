# Analisando Planilhas do Excel

```python
import pandas as pd
```

```python
df1 = pd.read_excel('Aracaju.xlsx')
df2 = pd.read_excel('Fortaleza.xlsx')
df3 = pd.read_excel('Natal.xlsx')
df4 = pd.read_excel('Recife.xlsx')
df5 = pd.read_excel('Salvador.xlsx')
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
df['LojaID'] = df['LojaID'].astype('object')
```

```python
df['Vendas'].fillna(df['Vendas'].mean(), inplace=True)
```

```python
df['Vendas'].mean()
```

```python
df.dropna(inplace=True)
```

```python
df.dropna(how='all', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Receitas'] = df['Vendas'].mul(df['Qtde'])
```

```python
df.head()
```

# Trabalhando com Datas

```python
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.groupby(df['Data'].dt.year)['Receitas'].sum()
```

```python
df['Ano_Venda'] = df['Data'].dt.year
```

```python
df['mes_venda'], df['dia_venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df['Data'].min()
```

```python
df['Data'].max()
```

```python
df['diferenca_dias'] = df['Data'] - df['Data'].min()
```

```python
df['trimestre_venda'] = df['Data'].dt.quarter
```

```python
vendas_marco_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
df.sample(200)
```

```python
vendas_marco_19
```

```python
import matplotlib.pyplot as plt
df['Cidade'].value_counts().plot.bar(title='total vendas por cidade')
plt.xlabel('Cidade')
plt.xlabel('total vendas')
```

```python
df.groupby(df['mes_venda'])['Qtde'].sum().plot(title = 'total produtos vendidos por mês')
plt.xlabel('Mês')
plt.xlabel('Total de Produtos vendidos')
plt.legend()
```
